# Simulating Colony Barcode Counts

Given a number of spatial locations for beads and bacterial colonies, we want to generate a bacterial barcodes count matrix for the beads. The barcode counts going to each bead are assigned according to a Gaussian kernel.

```python
import numpy as np
import pandas as pd
import gc
import anndata
import math
import matplotlib.pyplot as plt

import sys
sys.path.append("../functions/")
import simulate_barcodes
import simulate
import utils
```

```python
DATA_PATH = '../../data/'
```

```python
# Define parameters for the run
SEED = 42 # Random seed for reproducibility
RNG = np.random.default_rng(SEED)
NUM_PROC = 16

BEADS_PER_CM2 = 10**5
COLONIES_PER_CM2 = 10**4
SIGMA_PIX = 3   # 3: 1113 rpb
AREA_CM2 = 1

BEADS = int(BEADS_PER_CM2 * AREA_CM2)
COLONIES = int(COLONIES_PER_CM2 * AREA_CM2)
SAVING = False

# Factor by which colonies should extended past beads, this reduces edge effects where beads along the
# edge get fewer reads
COLONY_EXT = 0.05

BARCODE_OVERLAP = True
SUBSAMPLE_READS = True
READS_PER_BEAD = 1000 # Average number of reads per bead for the read sub-sampling
n = np.floor(np.sqrt(BEADS))
SIGMA = SIGMA_PIX/n  # Standard deviation for the Gaussian modelling diffusion around a colony
AMPLITUDE = 200 # The number of counts if a colony is directly on top of a bead
STDEV_NOISE = 5 # Magnitude of noise added to diffused colony barcode count

# Generate random colonies on a unit square and beads on a grid
y,x = np.meshgrid(np.arange(0,n), np.arange(0,n))
y = y.flatten()
x = x.flatten()
beads_df = np.array([x/max(x), y/max(y)]).T
colonies_df = RNG.random((int((COLONIES)*((1+2*COLONY_EXT)**2)), 2))*(1 + 2*COLONY_EXT) - COLONY_EXT
```

```python
if SAVING:
    np.savetxt(DATA_PATH + 'spatial-coordinates/%sB.tsv'%utils.human_format(BEADS), BEADS, delimiter='\t')
    np.savetxt(DATA_PATH + 'spatial-coordinates/%sC.tsv'%utils.human_format(COLONIES), COLONIES, delimiter='\t')
```

```python
beads_df = pd.DataFrame(data=beads_df, columns=['x', 'y'])
colonies_df = pd.DataFrame(data=colonies_df, columns=['x', 'y'])
```

```python
# Shows one of the point cloud dfs (beads or colonies)
def show_points(points_df, title, n_points=10000):
    # Determine how many points we are plotting
    n_points = np.min([n_points, points_df.shape[0]])
    
    plt.figure(figsize=(10, 10))
    plt.title(title)
    
    inds = list(points_df.index.copy())
    np.random.shuffle(inds)
    inds_to_plot = inds[:n_points]
    plt.scatter(points_df.loc[inds_to_plot, 'x'], points_df.loc[inds_to_plot, 'y'])
```

```python
show_points(beads_df, 'Beads')
show_points(colonies_df, 'Colonies')
```

### Read Data and Setup Params

```python
M = colonies_df.shape[0]

SAVE_PATH = DATA_PATH + 'adata/BC_distances_%sB_%sC_sigma%g'%(utils.human_format(BEADS),
                                                              utils.human_format(COLONIES),
                                                              SIGMA_PIX)
if not BARCODE_OVERLAP: SAVE_PATH += "_no-overlap"
if not SUBSAMPLE_READS: SAVE_PATH += "_no-read-subsampling"
SAVE_PATH += ".h5ad"
```

### Convert coordinates to counts with noise

First, we want to convert bead and colony coordinates to counts by applying a gaussian function on distance. Then, we want to add some normally distributed noise to the barcode counts.

```python
#Get a matrix of euclidean distances between the beads and colonies
beads_coords = beads_df.to_numpy()
colonies_coords = colonies_df.to_numpy()
```

```python
print("Build the read count matrix...")
colony_counts = simulate_barcodes.get_colony_counts(SIGMA, AMPLITUDE, beads_coords, colonies_coords, stdev_noise=STDEV_NOISE,
                                                    num_proc=NUM_PROC)
```

```python
print("Count matrix shape:",colony_counts.shape)
print("Average NNZ/row:",colony_counts.nnz/colony_counts.shape[0])
print("Average reads/bead:",colony_counts.sum(1).mean())
```

```python
# Make adata
adata = anndata.AnnData(X=colony_counts, obs=beads_df, var=colonies_df, dtype=np.int64)
```

### Barcode Overlap

We will have some proportion of barcodes overlap. We can add this proportion according a known distribution of barcodes.

```python
if BARCODE_OVERLAP:
    # Load the read distribution
    reads = pd.read_csv('../../data/experimental-data/pam_barcode_cnt.tsv', sep='\t', index_col=0)
    total_counts = reads['Count'].sum()
    reads['Dist'] = reads['Count']/total_counts
    # Introduce barcode overlaps
    print("Simulate barcode overlap...")
    new_counts = simulate.add_overlap_to_counts(adata.X, reads['Dist'])
    adata = anndata.AnnData(X=new_counts.copy(), obs=adata.obs.copy(), dtype=np.int64)
    print("New count matrix shape:", new_counts.shape)
```

## Perform Down-sampling
 
We downsample to a desired number of reads. We note that when data becomes very large, we may get overflow issues when downsampling. To counter this, we stricttly enforce the int64 dtype and also plot diagnostics before and after downsampling.

```python
def show_counts_dist(adata, title):
    plt.figure(figsize=(5, 5))
    plt.suptitle('Distribution of counts')
    
    # Plot the distribution of the original counts
    plt.title(title)
    plt.hist(adata.X.sum(axis=1, dtype=int))
```

```python
if SUBSAMPLE_READS:
    print("Downsample reads...")
    show_counts_dist(adata, title='Before')
    simulate.downsample_reads(adata, [READS_PER_BEAD], inplace=True)
    show_counts_dist(adata, title='After')
```

### Write Output as AnnData

```python
print("Write the anndata file.")
adata.write(SAVE_PATH)
```
